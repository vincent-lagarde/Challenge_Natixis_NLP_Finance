### Pre processing

```python

def suppr_footnotes(text):
    '''Fonction pour supprimer les footnotes d'un texte (rÃ©fÃ©rences biblio, etc)  (Ã  vÃ©rifier si permet de supprimer toutes les footnotes)'''
    
    try:
        txt = text.split('References', maxsplit=1)
        if len(txt)>1:
            return txt[0]
        else:
            1/0
    except:
        pass
    
    try:
        txt = text.split('Footnotes', maxsplit=1)
        if len(txt)>1:
            return txt[0]
        else:
            1/0
    except:
        pass
    try:
        txt = text.split('      [1]', maxsplit=1)
        if len(txt)>1:
            return txt[0]
        else:
            1/0
    except:
        pass
    
    try:
        txt = text.split('See also footnotes', maxsplit=1)
        if len(txt)>1:
            return txt[0]
        else:
            1/0
    except:
        pass

    try:
        txt = text.split(' References ', maxsplit=1)
        if len(txt)>1:
            return txt[0]
        else:
            1/0
    except:
        pass

    try:
        txt = text.split(' 1. ', maxsplit=1)
        if len(txt)>1:
            return txt[0]
        else:
            1/0
    except:
        return text
    
    try:
        txt = text.split('SEE ALSO', maxsplit=1)
        if len(txt)>1:
            return txt[0]
        else:
            1/0
    except:
        return text
    
    
```

```python
text_prepro = df_fed['text'].apply(func=suppr_footnotes)
```
